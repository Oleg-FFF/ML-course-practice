**1. Функція з параметром**

Створіть функцію `show_menu()`, яка приймає перелік страв, наприклад ['Піца', 'Паста', 'Салат'] і виводить меню ресторану в наступному вигляді:
```
=== МЕНЮ ===
1. Піца
2. Паста
3. Салат
=============
```

Викличте цю функцію на списку ["Суп грибний", "Котлета по-київськи", "Лимонад"].

````python
def show_menu(menu: list) -> None:
    """
    Show all dishes in menu.
    
    :param menu: list of dishes
    :return: None
    """
    
    print('=== МЕНЮ ===')
    for n, dish in enumerate(menu, 1):
        print(f'{n}. {dish}')
    print('============')
    
menu = ['Суп грибний', 'Котлета по-київськи', 'Лимонад']
show_menu(menu)
````

**2. Функція з параметром**

Створіть функцію `print_square(n)`, яка виводить квадрат числа.

Наприклад: `print_square(5)` → "Квадрат числа 5 = 25"

Викличте її з різними числами.

````python
def print_square(n: int) -> None:
    """
    Print square of given number.
    
    :param n: number
    :return: None
    """
    
    print(f'Квадрат числа {n} = {n ** 2}')
    
print_square(3)
print_square(6)
print_square(34)
````

**3. Функція з поверненням значення**

Створіть функцію `multiply(a, b)`, яка повертає добуток двох чисел.

Викличте функцію на власноруч заданих значеннях та виведіть результат.


````python
def multiply(a: int, b: int) -> int:
    """
    Return sum of given numbers.
    
    :param a: number a
    :param b: number b
    :return: sum
    """
    
    return a * b

print(multiply(3, 5))
print(multiply(67, 3))
print(multiply(21, 6))
````

**4. Функція пошуку максимумального значення**

Створіть функцію `find_max(numbers)`, яка:
- Приймає список чисел
- Повертає найбільше число в списку

Протестуйте на списку `[15, 42, 8, 23, 56, 19]`.


````python
def find_max(numbers: list) -> int:
    """
    Find max number in given list.
    
    :param numbers: list of numbers
    :return: max number
    """
    max_number = numbers[0]
    
    for n in numbers:
        if n > max_number:
            max_number = n
    return max_number

print(find_max([15, 42, 8, 23, 56, 19]))
````

**5. Функція з умовами**

Створіть функцію `check_grade(grade)`, яка:
- Приймає оцінку (число)
- Повертає "Здав" якщо grade >= 60
- Повертає "Не здав" якщо grade < 60

Протестуйте на різних оцінках.


````python
def check_grade(grade: int) -> str:
    """
    Check if exam is passed.
    
    :param grade: 
    :return: result
    """
    
    return 'Здав' if grade >= 60 else 'Не здав'

print(check_grade(4))
print(check_grade(78))
print(check_grade(60))
print(check_grade(-3))
````
